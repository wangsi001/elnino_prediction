```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import MultipleLocator
import matplotlib.patches as mpatches
from scipy.stats import pearsonr

metrix_year=[]
metrix_mae=[]
metrix_rmse=[]
metrix_pc=[]

for mae_rmse in range(len(all_prediction)):
    print (all_actual[mae_rmse])
    print (all_prediction[mae_rmse])
    # MAE
    Year=2008+mae_rmse
    print ("{} Year：".format(Year))
    
    mae_values=mean_absolute_error(all_actual[mae_rmse],all_prediction[mae_rmse])
    print ("mae_value is {0}".format(mae_values))
    
    # RMSE
    rmse_values=np.sqrt(mean_squared_error(all_actual[mae_rmse],all_prediction[mae_rmse]))
    print ("rmse_value is {0}".format(rmse_values))

    # IC
    pccs = pearsonr(all_actual[mae_rmse],all_prediction[mae_rmse])
    print ("Pearson correlation coefficient is {0}".format(pccs[0]))
    
    metrix_year.append(str(Year))
    metrix_mae.append(np.around(mae_values,2))
    metrix_rmse.append(np.around(rmse_values,2))
    metrix_pc.append(np.around(pccs[0],2))
    
metrix_all=[]
metrix_all.append(metrix_year)
metrix_all.append(metrix_mae)
metrix_all.append(metrix_rmse)
metrix_all.append(metrix_pc)
```
