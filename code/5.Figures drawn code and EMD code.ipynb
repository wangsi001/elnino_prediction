```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates
file_name_emd="ONI"
file_name="ONI_whole"
solo_signal=[]
all_signals=[]

with open("{0}.txt".format(file_name_emd)) as file:
    for lines in  file:
        every_signal=[]
        line=lines.split(",")
        for element in line:
            every_signal.append(float(element))
        all_signals.append(every_signal)

with open("{0}.txt".format(file_name)) as file:
    for lines in  file:
        every_signal=[]
        line=lines.split(",")
        for element in line:
            every_signal.append(float(element))
        solo_signal.append(every_signal)

data={}
pd_date=pd.date_range('1/1/1950','12/1/2018',freq='MS')
data["date"]=pd_date

for i in range(len(all_signals)):
    index="f"+str(i)
    data[index]=all_signals[i]

df=pd.DataFrame(data)
print (df.shape)


fig,axes=plt.subplots(10,1,figsize=(12,18))


axes[0].plot(data["date"],solo_signal[0],linewidth=3)
axes[0].set_title("Original Data",weight="bold",fontsize=16)
axes[0].tick_params(labelsize=12,direction='out',width=1,length=6)
axes[0].xaxis.set_major_locator(mdates.YearLocator(base=8))

axes[1].plot(data["date"],df["f0"][::],linewidth=1)
axes[1].set_title("IMF1",weight="bold",fontsize=14)


axes[2].plot(data["date"],df["f1"][::],linewidth=1)
axes[2].set_title("IMF2",weight="bold",fontsize=14)

axes[3].plot(data["date"],df["f2"][::],linewidth=1)
axes[3].set_title("IMF3",weight="bold",fontsize=14)


axes[4].plot(data["date"],df["f3"][::],linewidth=1)
axes[4].set_title("IMF4",weight="bold",fontsize=14)

axes[5].plot(data["date"],df["f4"][::],linewidth=1)
axes[5].set_title("IMF5",weight="bold",fontsize=14)
#axes[5].set_ylabel('SST anomalies ()',fontsize=15,weight="bold",color="black")

axes[6].plot(data["date"],df["f5"][::],linewidth=1)
axes[6].set_title("IMF6",weight="bold",fontsize=14)

axes[7].plot(data["date"],df["f6"][::],linewidth=1)
axes[7].set_title("IMF7",weight="bold",fontsize=14)

axes[8].plot(data["date"],df["f7"][::],linewidth=1)
axes[8].set_title("IMF8",weight="bold",fontsize=14)

axes[9].plot(data["date"],df["f8"][::],linewidth=1)
axes[9].set_title("Res",weight="bold",fontsize=14)
axes[9].set_xlabel('Date',fontsize=18,weight="bold",color="black",labelpad = 10)

for i in range(10):
    axes[i].xaxis.set_tick_params(labelsize=12,direction='out',width=1,length=6)
    axes[i].yaxis.set_tick_params(labelsize=10,direction='out',width=1,length=6)
    axes[i].xaxis.set_major_locator(mdates.YearLocator(base=8))
    axes[i].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    
#fig.text(0.53, -0.01, 'Time (Year)', ha='center', va='center',fontsize=15,weight="bold",color="black")
fig.text(0.005, 0.53, 'Oceanic Nio Index ()', ha='center', va='center', rotation='vertical',fontsize=18,weight="bold",color="black")

plt.tight_layout()

fig.savefig("EMD_picture.png",dpi=300)
```
